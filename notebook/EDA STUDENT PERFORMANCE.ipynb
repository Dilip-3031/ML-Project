```python
## numerical column 
numeric_features = [feature for feature in df.columns if df[feature].dtype == 'int64' ]
print('Numerical features:-',numeric_features)

##categorical_fetaures
categorical_features = [feature for feature in df.columns if df[feature].dtype == object]
print('categorical features:- ',categorical_features)
```

```python
## adding column total score and average 
df['Total_score'] = df['math_score']+df['reading_score']+df['writing_score']
df['Average'] = df['Total_score']/3
df.head()
```

```python
reading_full = df[df['reading_score']==100]['Average'].count()
writing_full = df[df['writing_score']==100]['Average'].count()
math_full = df[df['math_score']==100]['Average'].count()

print('number of student with full marks in reading score',reading_full)
print('number of student with full marks in writing score',writing_full)
print('number of student with full marks in math score',math_full)
```

## EDA
visualize avg score distribution to make some conclusion
* histogram 
* kernel distribution function (kde)

```python
## average

plt.Figure(figsize=(15,7))
sns.histplot(data=df,x = 'Average',hue = 'gender',kde = True)
plt.show()
sns.histplot(data = df, x = 'Average',kde = True,color = 'green')
```

```python
## Total score
plt.figure(figsize=(15,7))
sns.histplot(x = 'Total_score',kde = True,color = 'green',data = df)
plt.show()
sns.histplot(data = df,x = 'Total_score',kde = True,hue = 'gender')
plt.show()
```

here we can see that female perform well than boys 

```python
plt.Figure(figsize=(25,6))
sns.histplot(data=df,x = 'Average',kde = True,hue='lunch')
plt.show()
sns.histplot(data=df[df.gender=='female'],x = 'Average',kde = True,hue='lunch')
plt.xlabel('Average(Female)')
plt.show()
sns.histplot(data = df[df.gender == 'male'],x = 'Average',kde = True,hue='lunch')
plt.xlabel('Average(Male)')
plt.show()
```

### Insights 
* standard lunch helps to perform well in exam 
* standard linch helps both male and female to perform well 

```python
plt.figure(figsize=(25,20))
sns.histplot(data=df,kde=True,x = 'Average',hue='parental_level_of_education')
plt.show()
sns.histplot(data=df[df.gender=='female'],x = 'Average',kde = True,hue='parental_level_of_education')
plt.xlabel('Average(Female)')
plt.show()
sns.histplot(data=df[df.gender=='male'],x = 'Average',kde = True,hue='parental_level_of_education')
plt.xlabel('Average(Male)')
plt.show()
```
